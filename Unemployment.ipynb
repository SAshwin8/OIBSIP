```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
Unemployment = pd.read_csv("Unemployment.csv.csv")
```

```python
plt.style.use('seaborn-whitegrid')
plt.figure(figsize=(8, 6))
sns.heatmap(Unemployment.corr())
plt.show()
```

```python
Unemployment.columns= ["Region","Date","Frequency",
               "Estimated Unemployment Rate","Estimated Employed",
               "Estimated Labour Participation Rate","Area"]
plt.title("Indian Unemployment")
sns.histplot(x="Estimated Employed", hue="Area", data=Unemployment)
plt.show()
```

```python
plt.figure(figsize=(12, 10))
plt.title("Indian Unemployment based on Region")
sns.histplot(x="Estimated Unemployment Rate", hue="Region", data=Unemployment)
plt.show()
```

```python
plt.figure(figsize=(10, 8))
plt.title("Indian Unemployment based on labour participation rate in each region")
sns.histplot(x="Estimated Labour Participation Rate", hue="Region", data=Unemployment)
plt.show()
```
